# Rencontre à 2 avions

## Face à face

```python
# Définition des paramètres de la simulation

d = 25.
tau = 120.
time_step = 100.
area_size = 400.

# Configuration initiale
position1 = np.array([0., area_size/2])
destination1 = np.array([area_size, area_size/2])
heading1 = (destination1 - position1)/np.linalg.norm((position1 - destination1))

position2 = np.array([area_size, area_size/2])
destination2 = np.array([0., area_size/2])
heading2 = (destination2 - position2)/np.linalg.norm((position2 - destination2))

ac1 = Aircraft(position1, heading1, destination1)
ac2 = Aircraft(position2, heading2, destination2)

aircrafts = [ac1, ac2]

# Lancement de la situation

sim = Simulation(aircrafts, d, tau, area_size=area_size)
sim.run()
```

## Rencontre à 45 degrés

```python
# Définition des paramètres de la simulation

d = 25.
tau = 120.
time_step = 3.
area_size = 400.

# Configuration initiale
position1 = np.array([0., 0.])
destination1 = np.array([area_size, area_size])
heading1 = (destination1 - position1)/np.linalg.norm((position1 - destination1))

position2 = np.array([area_size, 0.])
destination2 = np.array([0., area_size])
heading2 = (destination2 - position2)/np.linalg.norm((position2 - destination2))

ac1 = Aircraft(position1, heading1, destination1)
ac2 = Aircraft(position2, heading2, destination2)

aircrafts = [ac1, ac2]

# Lancement de la situation

sim2 = Simulation(aircrafts, d, tau, area_size=area_size)
sim2.run()
```

# Rencontre à N avions

## Rencontre avec avions positionnés sur un arc de cercle avec destinations diamétralement opposées

```python
# Définition des paramètres de la simulation

N = 5
d = 10.
tau = 240.
time_step = 3.
area_size = 500.

# Configuration initiale
aircrafts = []

for i in range(N):
    angle = 5*np.pi/8 + i*(3*np.pi/(4*N))
    position = (area_size/2)*np.array((1-np.cos(angle + np.pi), 1-np.sin(angle + np.pi)))
    destination = (area_size/2)*np.array((1-np.cos(angle), 1-np.sin(angle)))
    heading = (destination - position)/np.linalg.norm((position - destination))
    
    aircrafts.append(Aircraft(position, heading, destination))

# Lancement de la situation

sim3 = Simulation(aircrafts, d, tau, area_size=area_size)
sim3.run()
```

```python
# Définition des paramètres de la simulation
import pdb
N = 5
d = 25.
tau = 60.
time_step = 5.
area_size = 500.

def avg_conflict():

    # Configuration initiale
    aircrafts = []
    coords = []

    for i in range(N):
        run = True
        while run:

        #position_edge, destination_edge = rd.choice(['UP', 'RIGHT', 'DOWN', 'LEFT']), rd.choice(['UP', 'RIGHT', 'DOWN', 'LEFT'])
            edge = ['UP', 'RIGHT', 'DOWN', 'LEFT']
            position_edge = rd.choice(edge)
            edge.remove(position_edge)
            destination_edge = rd.choice(edge)

            position_coord, destination_coord = rd.uniform(0,area_size), rd.uniform(0,area_size)

            position = np.array( np.array((position_coord, area_size))*(position_edge=='UP')
                                + np.array((area_size, position_coord))*(position_edge=='RIGHT')
                                + np.array((position_coord, 0.))*(position_edge=='DOWN')
                                + np.array((0., position_coord))*(position_edge=='LEFT'))

            destination = np.array( np.array((destination_coord, area_size))*(destination_edge=='UP')
                                + np.array((area_size, destination_coord))*(destination_edge=='RIGHT')
                                + np.array((destination_coord, 0.))*(destination_edge=='DOWN')
                                + np.array((0.,destination_coord))*(destination_edge=='LEFT'))
            if aircrafts:

                for aircraft in aircrafts:
                    if np.linalg.norm(position - aircraft.position) < d or np.linalg.norm(destination - aircraft.destination) < d:
                        run = True

                        break

                    else:
                        run = False
            else:
                run = False


        heading = (destination - position)/np.linalg.norm((position - destination))

        aircrafts.append(Aircraft(position, heading, destination))

    # Lancement de la situation
    sim3 = Simulation(aircrafts, d, tau, time_step=time_step)
    count_conflict = sim3.run(display=False, draw=False)
    #pdb.set_trace()
    return count_conflict/N
```
